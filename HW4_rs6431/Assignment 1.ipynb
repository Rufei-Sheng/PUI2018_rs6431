# 1. Normal distribution

```python
ns = 100     #number of samples in a distribution
my_mean = 10 #set up my mean as 10
my_sd = 3    #set up my standard deviation as 3

my_dict = {}
my_dict['normal'] = {} #set up key named "normal" with its value in a dict

for i in range (ns):
    diff_size = np.random.rand()*1990 + 10 #form different size between 10 and 2000
    my_dict['normal'][i] = np.random.normal(my_mean, my_sd, size=diff_size) #[i]assign different size number as an array into the ith value of 'normal' key

ax_n = pl.figure(figsize = (10,8)).add_subplot(111)
my_dict['normal']['sp_mean'] = {} #assign each sample's mean to 'sp_mean'
for k in my_dict['normal']:
    if not type(k) == str:
        sp_size = len(my_dict['normal'][k])
        my_dict['normal']['sp_mean'][k] = my_dict['normal'][k].mean()
        ax_n.plot(sp_size, my_dict['normal']['sp_mean'][k], 'o')
        
ax_n.set_title('Sample mean Vs Sample size in Normal Distribution', fontsize=15)
ax_n.set_xlabel('Sample Size', fontsize=15)
ax_n.set_ylabel('Sample mean', fontsize=15)
ax_n.plot([0, 2000], [my_mean, my_mean], '-', c='black')
```

Figure 1: This figure describes the sample mean aganist Sample size in Normal Distribution. Cetral limit theorm argues that when the sample size approches to infinite, the sample mean approaches a Normal (Gaussian) distribution, so the sample mean approches to the population mean. Through the figure above, the sample mean is randomly spread around the population mean. When size is small, the mean of sample spread loosely and even have some points far away from the population mean. On contrast, when sample size become larger (closer to 2000), the sample mean is closer to popluation mean(set as 10).

# Extral credit: FIT a gaussian to the distribution

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

all_mean = []
for i in range(0, len(my_dict)):
    sp_meani = list(my_dict[list(my_dict)[i]]['sp_mean'].values())
    all_mean += sp_meani

plt.figure(figsize=(10,8))
plt.hist(all_mean, bins=25, density=True, alpha=0.6, color='orange')
    
mu, std = norm.fit(all_mean) #fit a gaussian distribution to the data
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, linewidth=2)
plt.title("Gaussian fit results: mu = %.2f,  std = %.2f" % (mu, std), fontsize=15)
plt.xlabel('All sample means', fontsize=15)
```

Figure 7: this figure fit the Gaussian distribution to the histogram of all sample means distrbituion. After fitting, the mean and standard decviation for all sample means are 10.01 and 0.27, respectively.
