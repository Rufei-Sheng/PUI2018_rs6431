# Task 4: Fourier transform
**Several stations show a prominent annual periodicity. Identify the 4 stations (indentify them by the index of their location in the data cube) that show the most prominent periodic trend on an annual period (52 weeks). (Can you figure out what the periodic peak in rides is due to?)**

 from the numpy package you will nees only the 
np.fft.rfft() to take the fourier transform

and

np.fft.rfftfreq() to create a meaningful x axis 

`````python
mta.shape
`````

`````python
mta_bystation = np.nansum(mta, axis=1)
mta_bystation.shape
`````

`````python
#use 194 week as index, each station as features
mta_bystationT = pd.DataFrame(mta_bystation).T
mta_bystationT.head()
`````

`````python
fig = pl.figure(figsize=(10,8))
ax = fig.add_subplot(111)
N = len(mta_bystationT)
for i in range(mta_bystationT.shape[1]):
    f = np.abs(np.fft.rfft(mta_bystationT[i]))**2
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), 'o', ms=10)
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')
ax.set_xticklabels(['%s'%(1/f) for f in ax.get_xticks()])
pl.xlabel('period (week)', fontsize=20)
pl.ylabel('Power', fontsize=20)
pl.title('MTA swipes power spectrum for all stations', fontsize=20);
`````

**Figure 1:** The power spectrum of the number of swipes for all MAT stations. Some stations show prominent periodic trend.

### The 4 station that shows the most prominent periodic trend are:

`````python
N = len(mta_bystationT)
f_max = []
stationid = []
period = []

for i in range(mta_bystationT.shape[1]):
    f = np.abs(np.fft.rfft(mta_bystationT[i]))**2
    if f[1:].max() > 0.4*1e13:
        ith = np.where(f[1:] == f[1:].max())[0] + 1
        week = 1.0/(np.fft.rfftfreq(N, 1.0)[ith])
        stationid.append(i)
        f_max.append(f[1:].max())
        period.append(week[0])
        
data = {'stationID': stationid, 'period':period, 'f_max':f_max}
df = pd.DataFrame.from_dict(data)
df = df[df['period']<52].sort_values(by='f_max', ascending=False).reset_index(drop=True)
df
`````

`````python
fig = pl.figure(figsize=(10,8))
ax = fig.add_subplot(111)
for i in range(4):
    f = np.abs(np.fft.rfft(mta_bystationT[df.stationID[i]]))**2
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), 'o', ms=10, label = 'station %s'%df.stationID[i])
    ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')
ax.legend()
ax.set_xticklabels(['%s'%(1/f) for f in ax.get_xticks()])
ax.set_xlabel('period (week)', fontsize=20)
ax.set_ylabel('Power', fontsize=20)
ax.set_title('The 4 station that shows the most prominent periodic trend', fontsize=15)
`````

** Figure 2**: The 4 stations that have the prominent periodic trend are station 0, 194, 150, and 327 from orginial dataset. They all have the period as 48.5 weeks. 

### Visualize the four stations:

`````python
pl.figure(figsize=(15,8))
for i in range(4):
    mta_bystation.iloc[df.stationID[i]].plot(label='Station %s'%df.stationID[i])
    pl.legend(fontsize=10)
plt.title('Identify 4 stations with prominent periodic trend', fontsize=20)
plt.xlabel('194 weeks since 05/21/2010', fontsize=15)
plt.ylabel('number of swipes sumed by station', fontsize=15)
`````

** Figure 3**: Visualize the number of swipes from the 4 stations (station 0, 194, 150, and 327) that have the prominent periodic trend. The pattern of the number of swipes indeed change together. They all have the period as 48.5 weeks. 48 weeks roughly equal to 1 year, which indicates that the number of swipes has some stable change pattern annually.

#### Note: Choose to use all time stamps to do the clustering

`````python
mta_rs = mta.reshape(600*23,194)
mta_rs.shape
`````

`````python
#apply the label
label = np.asarray([np.arange(0,mta.shape[1])]*mta.shape[0]).flatten() 
label[:26]
`````

`````python
label_df = pd.DataFrame(np.transpose([label]))
label_df.rename(columns={0:'ridetype'}, inplace=True)
label_df.shape
`````

`````python
label_df.head()
`````

`````python
pd.DataFrame(mta_rs).head()
`````

`````python
tot_df = pd.merge(label_df, pd.DataFrame(mta_rs), left_index=True, right_index=True)
tot_df.dropna(inplace=True)
tot_df.reset_index(inplace=True, drop=True)
tot_df.head()
`````

`````python
tot_df.shape
`````

`````python
label = tot_df['ridetype']
mta_tot = tot_df.drop(['ridetype'], axis=1)
`````

`````python
#Standaridize the time series data:
scaler = preprocessing.StandardScaler().fit(mta_tot.T)
mta_totT = pd.DataFrame(scaler.transform(mta_tot.T))
mta_tot = mta_totT.T
`````

`````python
#test the mean=0, std=1
mta_tot.iloc[0].mean(), mta_tot.iloc[0].std()
`````

`````python
from sklearn.cluster import KMeans
`````

`````python
kmeans = KMeans(n_clusters=23, max_iter=600)
kmeans.fit(mta_tot)
`````

`````python
cluster = pd.DataFrame(kmeans.labels_).rename(columns={0:'cluster'})
tot_cluster = pd.merge(cluster, tot_df, left_index=True, right_index=True)
tot_cluster.head()
`````

`````python
tot_cluster['cluster'] = tot_cluster['cluster'].astype(int)
for i in range(24):
    df_sub = tot_cluster[tot_cluster['cluster']==i]
    print('The ridetype in %s cluster are:'%i)
    print(df_sub['ridetype'].value_counts())
#tot_cluster.sort_values(by = 'cluster').head(10)
`````

**Interpretation:** I choose 23 clusters aim at simulating the number of card type. After clustering, comparing the categorical results with the true ridetype. However, it is obvious that each cluster contains mass ridetype. Therefore, the clusters seems don't reflect the card type.
